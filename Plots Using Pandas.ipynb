```python
ik=pa.read_csv('https://raw.githubusercontent.com/venky14/Machine-Learning-with-Iris-Dataset/master/Iris.csv')
```

```python
ik
```

```python
ik1=ik.drop('Id',axis=1)   #   Removing unnecessary column /  outlier
```

```python
ik1.plot(figsize=(17,9))  # it's lineplot
```

```python
ik1.plot(kind='bar',figsize=(17,9))  
# This is barplot, kind indicates type of plot that we want(line is default,bar,pie,scatter,area,box..are other plots)
```

```python
ik1.iloc[[7]].plot(kind='bar',figsize=(17,9)) 
# iloc function fetches the data within given range.Here(1 value)
```

```python
ik1.iloc[7:24].plot(kind='bar',figsize=(17,9))
# iloc function fetches the data within given range.Here(7 to 24) vertically.
```

```python
ik1.iloc[9:18].plot(kind='barh',figsize=(17,9))
# iloc function fetches the data within given range,horizontally just by putting kind=barh instead of bar.
```

```python
ik1.iloc[5:21]
```

```python
ik1.iloc[3:47].plot(kind='hist',figsize=(9,6))  
```

```python
ik1['SepalWidthCm'].plot(kind='hist',figsize=(8,5),orientation='horizontal')
#  ik1['SepalWidthCm'].plot(kind='hist',figsize=(8,5))
```

```python
 ik1.hist(figsize=(16,9),color='blue' , alpha= .5) 
# when we take histogram as function ,it plots graph for each column seperately
# Here alpha is used to increase or decrease the contrast of graphs.
```

```python
ik1.plot(kind='box',figsize=(18,9),color={'boxes':'b','whiskers':'b','medians':'r','caps':'r'})
#  In color boxes means the box ,
#  median means 50th percentile line inside boxes, 
#  whiskers means intersecting line along boxes at 90 degrees
#  caps are lines parallel to the boxes in red
```

```python
ik1.plot(kind='area',figsize=(17,9),alpha=0.5)
```

```python
ik1.plot(kind='area',figsize=(17,9),alpha=0.5 , stacked = False)
# when stacked is false,It shows the transparency of graphs plotted one behind.
```

```python
ik1.plot(kind='scatter',x='SepalLengthCm',y='SepalWidthCm',figsize=(11,5),alpha=0.5)
# scatter plot shows the relation between 2 columns visually
```

```python
ik1.plot.hexbin(x='SepalLengthCm',y='SepalWidthCm',figsize=(11,5),gridsize=25)
```
